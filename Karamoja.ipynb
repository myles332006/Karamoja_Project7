This project aims to develop a reliable food security tool to estimate and identify the hotspots of crop production across the area of Karamoja.

# OBJECTIVES:
1) What is the average Sorghum and Maize yield across the subcounties and districts of Karamoja?

2) Distinguish the area of Maize and Sorghum across Karamoja.

3) How are the Karamoja districts performing in terms of crop production?

4) Which areas are at risk of food insecurity?

5) Which subcounty is the most vulnerable for crop production?

# DATA UNDERSTANDING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the csv
df = pd.read_csv('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv')
df
```

```python
df.isnull().sum() #checking for null values
```

Data is clean, now let's visualize

# 1) Average Sorghum and Maize Yield Per Hectare

District Level

```python
avg_yield_district = df.groupby('DISTRICT_NAME')[['M_Yield_Ha', 'S_Yield_Ha']].mean().reset_index()
avg_yield_district
```

SUBCOUNTY LEVEL

```python
avg_yield_subcounty = df.groupby('SUBCOUNTY_NAME')[['M_Yield_Ha', 'S_Yield_Ha']].mean().reset_index()
avg_yield_subcounty
```

With this, the NGO can:
1) Focus more on sorghum productivity since it lags behind maize, despite being a key staple.

2) Keep an eye on vulnerable hotspots in which can be used to increase crop yields.

# 2) Comparison of Area for Maize and Sorghum

```python
area_district = df.groupby('DISTRICT_NAME')[['M_Area_Ha', 'S_Area_Ha', 'Crop_Area_Ha']].sum().reset_index()
area_district
```

```python
area_district.plot(x='DISTRICT_NAME', y=['M_Area_Ha', 'S_Area_Ha', 'Crop_Area_Ha'], kind='bar', figsize=(12, 6))
plt.title('Comparison of Area of Maize and Sorghum')
plt.xlabel('District')
plt.ylabel('Area (ha)')
plt.xticks(rotation=45)
plt.show()
```

With this the NGO can get a clearer picture of the amount of land used for these crop productions by the highlighted districts for each crop.

# 3) Crop Performance per District

```python
avg_yield_district.plot(x='DISTRICT_NAME', y=['M_Yield_Ha', 'S_Yield_Ha'], kind='bar', figsize=(12, 6))
plt.title('Average Sorghum vs Maize Yield per District')
plt.xlabel('District')
plt.ylabel('Yield (kg/ha)')
plt.xticks(rotation=45)
plt.show()
```

With this the NGO can highlight the best and worst performing district for each crop.

# 4) Hotspot Identification

```python
plt.figure(figsize=(10,6))
sns.heatmap(avg_yield_district.set_index("DISTRICT_NAME"), annot=True, cmap="RdYlGn")
plt.title("Hotspot Identification")
plt.show()
```

With this information, The NGO can:
1) Compare Sorghum vs Maize yields side by side for each district.

2) Quickly spot underperforming districts. (The dark red squares)

3) See which crops are doing better in different areas.

# 5) Crop Production vs Population

```python
df["Total_Prod"] = df["S_Prod_Tot"] + df["M_Prod_Tot"]

plt.figure(figsize=(10,6))
plt.scatter(df["POP"], df["Total_Prod"], alpha=0.6)
plt.xlabel("Population")
plt.ylabel("Total Production (kg)")
plt.title("Production vs Population by Subcounty")
plt.show()
```

Findings:
1. Sorghum yields are generally lower than maize yields across districts.
2. The best-performing district shows significantly higher productivity, while some subcounties are still lurking behind.
3. Population is unevenly distributed; some subcounties with large populations still show low production, creating food insecurity risks.
4. Choropleth maps reveal clear spatial disparities, with certain districts consistently underperforming.
5. NGOs can prioritize interventions in hotspot subcounties where both yield and total production are low but population is high.
